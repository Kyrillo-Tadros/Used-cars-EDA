```python
def update_fuel_type(row):
    if 'Other' in row['fuel_type'] and row['range'] < 100:
        return 'Hybrid'
    elif 'Other' in row['fuel_type'] and row['range'] >= 100:
        return 'Electric'
    elif 'Other' in row['fuel_type'] and row ['fuel_consumption_l_100km'] > 7.5:
        return 'Petrol'
    elif 'Other' in row['fuel_type'] and row ['fuel_consumption_l_100km'] > 3.5:
        return 'Diesel'
    elif 'Other' in row['fuel_type'] and row ['fuel_consumption_l_100km'] > 0:
        return 'Hybrid'
    elif 'Other' in row['fuel_type'] and row ['fuel_consumption_l_100km'] == 0:
        return 'Electric'
    else:
        return row['fuel_type'] 

cars['fuel_type'] = cars.apply(update_fuel_type, axis=1)
```

```python
def fill_nans_by_fuel_type(df):
    fuel_types = ['Petrol', 'Diesel', 'Hybrid', 'LPG', 'CNG', 'Diesel Hybrid', 'Electric', 'Ethanol', 'Hydrogen']
    
    for fuel_type in fuel_types:
        mean_value = df.loc[df['fuel_type'] == fuel_type, 'fuel_consumption_l_100km'].mean()
        df.loc[(df['fuel_type'] == fuel_type) & (df['fuel_consumption_l_100km'].isnull()), 'fuel_consumption_l_100km'] = mean_value.round(1)
    
    return df

cars = fill_nans_by_fuel_type(cars)
```
